```python
import scipy.special as sp
```

```python
x = sp.jn_zeros(0,3)
y = sp.jn_zeros(1,3)
z = sp.jn_zeros(2,3)
(x,y,z)
```

```python
%pylab inline
```

```python
x = linspace (0,16,50000)
y0 = sp.jn(0,x)
y1 = sp.jn(1,x)
y2 = sp.jn(2,x)
```

```python
plot(x,y0,'b')
plot(x,y1,'g')
plot(x,y2,'r')
grid()
```
