```python
data = loadtxt('sampled+ma0844az_1-1+_data.txt')
n = len(data[:,1])
m = len(data[1,:])
x = linspace(1,n,n)
fourier = empty([n,m], dtype=complex)
```

```python
for i in range(0,m):
    plot(x,data[:,i])
```

```python
from scipy.fftpack import fft, fftfreq, ifft

for i in range(0,m):
    fourier[:,i] = fft(data[:,i])
    
freq = fftfreq(n,1)
```

```python
plot(freq,log10(abs(fourier[:,1])**2))
yscale('log')
```

```python
for i in range(0,m):
    mayores = sort(abs(fourier[:,i]), axis=None)[-10:]
    fourier[:,i][abs(fourier[:,i])<mayores[0]] = 0
```

```python
plot(freq, abs(fourier[:,3])**2)
```

```python
new = empty([n,m], dtype=complex)
for i in range(0,m):
    new[:,i] = ifft(fourier[:,i])
```

```python
plot(x,real(new[:,5]), 'b')
plot(x,data[:,5], 'r')
```
